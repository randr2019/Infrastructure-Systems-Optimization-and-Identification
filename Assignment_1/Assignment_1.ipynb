## Transition probability for reconstruction

## forming Transition probability matrix

```python
P_tr=np.array([donot_mat,Rehab_mat,recon_mat])
```

## Reading action_cost file

```python
cost_df=pd.read_csv('C:\Graduate Courses\Spring 2020\CGN_5617\Assignment_1\cost.csv',header=None)
C=np.array(cost_df.values)
C
```

## Define user cost 

```python
#User Cost
U=np.zeros((10))
big_number=1000
U[7:10]=big_number
print("User_cost: ", U)

#Action
a=[0,1,2]
print("action :",a)

#discount factor
gamma=1.0
```

```python
v_y = U.copy()    ## intiliaze the long term cost for (t+1) or (100+1) year
v_y
```

```python
## values for different years and different action
v=np.zeros((101,10,3))
```

```python
## State
x=np.array(range(10))
x
```

```python
prediction_horizon=100
```

## Optimal cost and action estimation 

```python
## Optimal cost and action estimation 

for t in range(100,0,-1):
    print(t)
    for i in range (10):
        for k in a:
            C_l=0     #expected long term cost
            for j in range(10):
                if (t==prediction_horizon):
                    C_l= C_l + P_tr[k][i][j]*(U[i]) + P_tr[k][i][j]*(v_y[j])               # for 100 year use cost v_y 
                else :
                    C_l= C_l + P_tr[k][i][j]*(U[i]) + P_tr[k][i][j]*min(v[t+1][j][:])      # use minimum cost from the v table 
            
            print(C_l)
            
            v[t][i][k]=C[k][i]+gamma*C_l
            
            print(v[t][i][k])
```

```python
print(v.shape)

## cost for three differnt action for 1 to 100 year
## we don not need the year 0

V_t=np.array(v[1:101])

print(V_t.shape)
```

```python
V_t.shape
```

```python
V_re=V_t.reshape(100*10,3)
```

```python
V_re.shape
```

```python
## Estimation minimum cost
## Best action from he cost values

state=[]
action=[]
Optimal_cost=[]

for i in range (len(V_re)):
    action.append(np.argmin(V_re[:][i]))
    Optimal_cost.append(min(V_re[:][i]))
    state.append(list(range(9,-1,-1)))
```

```python
## 2D matrix in to 1D list

state=np.array(state).flatten()
```

## Creating a dataframe to save the results as a table

```python
df_result=pd.DataFrame(V_re,columns=["C1","C2","C3"])
df_result['state']=pd.DataFrame(state)
df_result['action']=pd.DataFrame(action)
df_result['optimal_cost']=pd.DataFrame(Optimal_cost)

df_result.head(10)
```

```python
## including year column

df_result['year']=1
i=0
j=0
while i <=1000:
    df_result['year'][i:i+10] = j+1
    #print(df_result['year'][i:i+10])
    i=i+10
    j=j+1
```
